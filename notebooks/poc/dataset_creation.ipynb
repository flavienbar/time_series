```python
import pandas as pd
import numpy as np
from datetime import date
from datetime import datetime
from pycoingecko import CoinGeckoAPI
```

# Coingecko API

## Connection

```python
cg = CoinGeckoAPI()
```

```python
coin = cg.get_coin_market_chart_by_id(id='bitcoin',vs_currency='usd',days='90')
coin
```

```python
coin.keys()
```

## Prices

```python
p_array = np.array(coin['prices'])
price = p_array[:,1]
```

## Dates

```python
def unix_to_datetime(unix_time):
    '''unix_to_datetime(1622505700)=> ''2021-06-01 12:01am'''a
    ts = int(unix_time/1000 if len(str(unix_time)) > 10 else unix_time) # /1000 handles milliseconds
    return datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %l:%M%p').lower()
```

```python
def datetime_to_unix(year, month, day):
    '''datetime_to_unix(2021, 6, 1) => 1622505600.0'''
    dt = datetime(year, month, day)
    timestamp = (dt - datetime(1970, 1, 1)).total_seconds()
    return timestamp
```

```python
time = [ unix_to_datetime(i[0]) for i in coin['prices'] ]
```

## Volume

```python
v_array = np.array(coin['total_volumes'])
volume = v_array[:,1]
```

## Create the Dataframe

```python
coingecko_raw = pd.DataFrame({'time':time, 'price':price, 'volume':volume})
coingecko_raw.head()
```
